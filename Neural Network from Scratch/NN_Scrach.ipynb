```python
import numpy as np

# =============================================================
# SIMPLE LINEAR MODEL FROM SCRATCH (NO HIDDEN LAYERS)
# -------------------------------------------------------------
# Goal:
#   Learn the function: y = 2x + 1
#
# Approach:
#   We use the simplest possible neural network:
#       Input (1) → Output (1)
#
#   This is just a linear model:
#       y_pred = w * x + b
#
#   And we manually implement:
#       - Forward pass
#       - Loss computation
#       - Backpropagation (gradients)
#       - Parameter updates
#
# Key Idea:
#   This is exactly what deep networks reduce to for linear problems.
# =============================================================

np.random.seed(42)

# =============================================================
# 1. DATASET
# =============================================================

X = np.array([
    [1.0],
    [2.0],
    [3.0]
])  # shape: (3,1)

Y = np.array([
    [3.0],
    [5.0],
    [7.0]
])  # shape: (3,1)

# =============================================================
# 2. INITIALIZE PARAMETERS
# =============================================================

# Single weight and bias
w = np.random.randn(1, 1)  # shape (1,1)
b = np.zeros((1, 1))       # shape (1,1)

lr = 0.01

# =============================================================
# 3. TRAINING LOOP
# =============================================================

epochs = 1000

for epoch in range(epochs):

    # ----------------------
    # FORWARD PASS
    # ----------------------
    # y_pred = w*x + b
    y_pred = X @ w + b   # (3,1)

    # ----------------------
    # LOSS (Mean Squared Error)
    # ----------------------
    loss = np.mean((y_pred - Y) ** 2)

    # ----------------------
    # BACKWARD PASS
    # ----------------------
    # dLoss/dy_pred
    dZ = 2 * (y_pred - Y) / X.shape[0]

    # Gradients
    dW = X.T @ dZ    # (1,1)
    db = np.sum(dZ, axis=0, keepdims=True)  # (1,1)

    # ----------------------
    # UPDATE
    # ----------------------
    w -= lr * dW
    b -= lr * db

    # ----------------------
    # LOGGING
    # ----------------------
    if epoch % 100 == 0:
        print(f"Epoch {epoch}, Loss: {loss:.6f}, w={w[0][0]:.4f}, b={b[0][0]:.4f}")

# =============================================================
# 4. FINAL RESULTS
# =============================================================

print("\nFinal learned parameters:")
print("w ≈", w)
print("b ≈", b)

print("\nPredictions:")
print(y_pred)

print("\nActual:")
print(Y)

# =============================================================
# 5. TEST ON NEW DATA
# =============================================================

X_test = np.array([[5.0]])

y_test_pred = X_test @ w + b

print("\nPrediction for x=5:", y_test_pred)
print("Expected: 11")
```
