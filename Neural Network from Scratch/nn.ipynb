# Neural Network From Scratch (1→4→4→1)

Learn **y = 2x + 1** using manual forward pass and backpropagation.


````python
import numpy as np
````

```markdown
# NEURAL NETWORK FROM SCRATCH (NO PYTORCH / TENSORFLOW)

## Goal:
Learn the function: y = 2x + 1

## Approach:
We build a small neural network:

Input (1) → Hidden (4) → Hidden (4) → Output (1)

And manually implement:
- Forward pass
- Loss computation
- Backpropagation (gradients)
- Parameter updates

## Key Idea:
The model does NOT "know" the equation.
It only learns from input-output pairs (X, Y).
```

````python
np.random.seed(42)  # For reproducibility (same random values each run)
````

```markdown
## 1. DATASET (Training Data)

Each row is one training example
X: input values (shape: [num_samples, num_features])
Y: target outputs (shape: [num_samples, 1])

We are teaching:
y = 2x + 1
```

````python
X = np.array([
    [1.0],
    [2.0],
    [3.0]
])  # shape: (3,1)

Y = np.array([
    [3.0],
    [5.0],
    [7.0]
])  # shape: (3,1)

print("X shape:", X.shape)
print("Y shape:", Y.shape)
````

```markdown
## 2. INITIALIZE PARAMETERS (WEIGHTS & BIASES)

Each layer has:
Z = XW + b

Shapes must align for matrix multiplication
```

````python
# Layer 1: (1 → 4)
W1 = np.random.randn(1, 4)
b1 = np.zeros((1, 4))

# Layer 2: (4 → 4)
W2 = np.random.randn(4, 4)
b2 = np.zeros((1, 4))

# Layer 3: (4 → 1)
W3 = np.random.randn(4, 1)
b3 = np.zeros((1, 1))

# Learning rate (controls how big each update step is)
lr = 0.01
````

```markdown
## 3. ACTIVATION FUNCTION (ReLU)

ReLU(x) = max(0, x)

Used to introduce non-linearity
```

````python
def relu(x):
    return np.maximum(0, x)


def relu_grad(x):
    """
    Gradient of ReLU:
    - 1 if x > 0
    - 0 if x <= 0
    """
    return (x > 0).astype(float)
````

```markdown
## 4. TRAINING LOOP

Each epoch does:
1. Forward pass (compute predictions)
2. Compute loss (error)
3. Backward pass (compute gradients)
4. Update weights
```

````python
epochs = 1000

for epoch in range(epochs):

    # =========================================================
    # FORWARD PASS
    # ---------------------------------------------------------
    # Data flows from input → output
    # =========================================================

    # Layer 1
    Z1 = X @ W1 + b1      # Linear transformation
    A1 = relu(Z1)         # Activation

    # Layer 2
    Z2 = A1 @ W2 + b2
    A2 = relu(Z2)

    # Layer 3 (output layer, no activation for regression)
    Z3 = A2 @ W3 + b3
    y_pred = Z3

    # =========================================================
    # LOSS COMPUTATION (Mean Squared Error)
    # ---------------------------------------------------------
    # Measures how far predictions are from actual values
    # =========================================================

    loss = np.mean((y_pred - Y) ** 2)

    # =========================================================
    # BACKWARD PASS (Backpropagation)
    # ---------------------------------------------------------
    # We compute gradients using chain rule
    # Starting from output → going backwards
    # =========================================================

    # ----------------------
    # Step 1: Output layer
    # ----------------------

    # dLoss/dy_pred
    dZ3 = 2 * (y_pred - Y) / X.shape[0]  # divide by batch size

    # Gradients for W3 and b3
    dW3 = A2.T @ dZ3
    db3 = np.sum(dZ3, axis=0, keepdims=True)

    # Propagate gradient backward to Layer 2
    dA2 = dZ3 @ W3.T

    # ----------------------
    # Step 2: Layer 2
    # ----------------------

    # Apply ReLU gradient
    dZ2 = dA2 * relu_grad(Z2)

    # Gradients for W2 and b2
    dW2 = A1.T @ dZ2
    db2 = np.sum(dZ2, axis=0, keepdims=True)

    # Propagate backward
    dA1 = dZ2 @ W2.T

    # ----------------------
    # Step 3: Layer 1
    # ----------------------

    dZ1 = dA1 * relu_grad(Z1)

    dW1 = X.T @ dZ1
    db1 = np.sum(dZ1, axis=0, keepdims=True)

    # =========================================================
    # UPDATE STEP (Gradient Descent)
    # ---------------------------------------------------------
    # Adjust weights in direction that reduces loss
    # =========================================================

    W3 -= lr * dW3
    b3 -= lr * db3

    W2 -= lr * dW2
    b2 -= lr * db2

    W1 -= lr * dW1
    b1 -= lr * db1

    # =========================================================
    # LOGGING (monitor training)
    # =========================================================

    if epoch % 100 == 0:
        print(f"Epoch {epoch}, Loss: {loss:.6f}")
````

```markdown
## 5. FINAL RESULTS
```

````python
print("\nFinal Predictions:")
print(y_pred)

print("\nActual Values:")
print(Y)
````
